```python
#Figure 2.2

import numpy
import matplotlib.pyplot as plt

w = numpy.logspace(-3, 1, 1000) 
s =1j*w
G = 5*numpy.exp(-2*s)/(10*s +1)
PG = numpy.unwrap(numpy.angle(G)*(180/numpy.pi),discont=180 )
GM = numpy.abs(G)

plt.figure('')
plt.subplot(2,1,1)
plt.title('BODE PLOT\n')
plt.loglog(w, GM)
plt.xlabel('Frequency in rad/s')
plt.ylabel('Magnitude')
plt.axhline(y = 1, linestyle = '--', color = 'orange')
plt.grid()
plt.subplot(2,1,2)
plt.semilogx(w, PG)
plt.xlabel('Frequency in rad/s')
plt.ylabel('Phase')
plt.axhline(y = -180, linestyle = '--', color = 'red')
plt.grid()
plt.savefig('Example 1', format= 'pdf')
plt.show()
```
